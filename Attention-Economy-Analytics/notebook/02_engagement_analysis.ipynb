```python
#1. Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2.Load & Inspect Data
df = pd.read_csv("../data/synthetic/attention_platform_monthly.csv")

df.info()
df.describe()
```

```python
#3.Attention Consumption Overview
# Average Daily Minutes by Platform
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x="platform",
    y="avg_daily_minutes",
    estimator="mean"
)
plt.title("Average Daily Attention Time by Platform")
plt.ylabel("Minutes per Day")
plt.xlabel("")
plt.show()
```

```python
#4.Session Depth vs Frequency (KEY ANALYSIS)
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=df,
    x="sessions_per_day",
    y="avg_session_duration_min",
    hue="category"
)
plt.title("Session Frequency vs Session Duration")
plt.xlabel("Sessions per Day")
plt.ylabel("Session Duration (min)")
plt.show()
```

```python
# 5.Notification Pressure Analysis
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x="platform",
    y="notifications_per_day"
)
plt.title("Notification Pressure by Platform")
plt.ylabel("Notifications per Day")
plt.xlabel("")
plt.show()
```

```python
# 6.Ads vs Retention Relationship
plt.figure(figsize=(7,5))
sns.regplot(
    data=df,
    x="ads_per_hour",
    y="monthly_retention_rate",
    scatter_kws={"alpha":0.3}
)
plt.title("Ad Load vs User Retention")
plt.xlabel("Ads per Hour")
plt.ylabel("Monthly Retention Rate")
plt.show()
```

```python
# 7.Age Group Attention Behavior
age_attention = df.groupby("user_age_group")["avg_daily_minutes"].mean()
age_attention.plot(kind="bar", figsize=(6,4))
plt.title("Average Daily Attention by Age Group")
plt.ylabel("Minutes per Day")
plt.xlabel("Age Group")
plt.show()
```
