```python
#1. Import Required Libraries
import pandas as pd
import numpy as np
```

```python
# Platforms and categories
platforms = {
    "Instagram": "Social",
    "YouTube": "Video",
    "Netflix": "Video",
    "Spotify": "Audio",
    "Coursera": "EdTech",
    "Google News": "News"
}
```

```python
months = pd.date_range(start="2023-01-01", periods=24, freq="ME")

data = []

np.random.seed(42)

for platform, category in platforms.items():
    for month in months:
        avg_minutes = np.random.uniform(20, 120)
        session_duration = np.random.uniform(5, 40)
        sessions = np.random.uniform(1, 6)
        notifications = np.random.uniform(1, 15)
        ads = np.random.uniform(0, 12)
        retention = np.random.uniform(0.6, 0.95)
        content_refresh = np.random.uniform(0.3, 0.9)

        data.append([
            platform,
            category,
            month.strftime("%Y-%m"),
            round(avg_minutes, 1),
            round(session_duration, 1),
            round(sessions, 1),
            round(notifications, 1),
            round(ads, 1),
            "Freemium",
            round(retention, 2),
            round(content_refresh, 2),
            np.random.choice(["18-24", "25-34", "35-44"])
        ])

df = pd.DataFrame(data, columns=[
    "platform", "category", "year_month",
    "avg_daily_minutes", "avg_session_duration_min",
    "sessions_per_day", "notifications_per_day",
    "ads_per_hour", "subscription_model",
    "monthly_retention_rate", "content_refresh_rate",
    "user_age_group"
])
df.head()
```

```python
#Save Dataset
df.to_csv("../data/synthetic/attention_platform_monthly.csv", index=False)
```
